```python
from  sympy import *
```

```python
x = Symbol("x")
```

```python
f, g = Symbol('f(x)'), Symbol('g(x)')
```

```python
funcF = x**3 - 50 * x
```

```python
funcG = -x**4 + 88 * x**2 - 241
```

```python
funcE = solve(funcF - funcG, x)
```

```python
absc = []
```

```python
for i in funcE:
    absc.append(round(i.n(10), 2))
```

```python
absc
```

```python
ordi = [funcF.subs(x, i) for i in absc]
```

```python
ordi
```

```python
for i, (x, y) in enumerate(zip(absc, ordi), start = 1):
    print('пересечение с осью абсцисс ', i, ' точки = (', x,',', y, ')', sep = '')
```
